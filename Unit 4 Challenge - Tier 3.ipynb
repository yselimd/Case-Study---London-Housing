**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# transposing the dataframe and renaming it as df.
df = properties.transpose()
#resetting the index of the dataframe and keeping the index as a column.
df.reset_index(inplace=True)
#setting the first row as column labels. 
df.columns = df.iloc[0]
#dropping the first row.
df=df.drop(df.index[0])
#Checking the first 5 rows of the dataframe.
df.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Renaming the columns (First and second columns as London_Borough and ID, respectively.)
df = df.rename(columns={df.columns[0]:'London_Borough',df.columns[1]:'ID'})
#Checking the first 5 rows of the dataframe.
df.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data?

```python
# Melting the data frame so that I will keep the column labels as Month except the "London_Borough" and "ID" and
# will rename the variable name as "Month" and value name as " Average_Price"
df_melt = pd.melt(df,id_vars = ["London_Borough","ID"],var_name='Month',value_name='Average_Price')
df_melt.head()
```
